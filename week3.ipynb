2. Special series regarding the definition of e: (1+1/n)^n, (1+1/n)^n+1, n^{1/n}

**Plot 1**: `e_n=(1+1/n)^n vs. f_n=(1+1/n)^(n+1)`

**Plot 2**: `f(x) = ln x / x (x>0) vs. g(x) = x^{1/x} (x>0)`

**Plot 3**: `y=g(e) vs. y=g(2) vs. y=g(3)`

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
n = np.arange(1, 50)

##################### Fill Here ###################
e_n = (1+1/n)**n
f_n = (1+1/n)**(n+1)
###################################################

plt.plot(n, e_n, 'b.-', label='e_n = (1 + 1/n)^n (Increases to e)')
plt.plot(n, f_n, 'r.-', label='f_n = (1 + 1/n)^(n+1) (Decreases to e)')
plt.axhline(y=np.e, color='k', linestyle='--', label='y = e (approx 2.718)')

plt.title('Plot 1: Sequences converging to e')
plt.xlabel('n')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
x = np.linspace(0.1, 15, 400)

##################### Fill Here ###################
f_x = np.log(x) / x # np.log is natural logarithm
g_x = x**(1/x)
###################################################

plt.plot(x, f_x, 'g-', label='f(x) = ln(x) / x')
plt.plot(x, g_x, 'm-', label='g(x) = x^(1/x)')
plt.axvline(x=np.e, color='r', linestyle='--', label=r'$x=e$ (local maximum)')

plt.title('Plot 2: f(x) and g(x) has the same shape')
plt.xlabel('x (x > 0)')
plt.ylabel('y')
plt.ylim(-2, 3)
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8, 6))

x_pos = np.linspace(0.5, 10, 400)
g_x_pos = x_pos**(1/x_pos)

############### Fill Here ###################
val_e = np.e**(1/np.e) # np.e is exponential
val_2 = 2**0.5
val_3 = 3**(1/3)
#############################################

plt.plot(x_pos, g_x_pos, 'm-', alpha=0.5, label='g(x) = x^(1/x)')

plt.axhline(y=val_e, color='r', linestyle='-', label=f'y = g(e) approx {val_e:.3f}')
plt.axhline(y=val_3, color='b', linestyle='--', label=f'y = g(3) approx {val_3:.3f}')
plt.axhline(y=val_2, color='g', linestyle='-.', label=f'y = g(2) approx {val_2:.3f}')

plt.plot([np.e, 2, 3], [val_e, val_2, val_3], 'ko')

plt.title('Plot 3: Maxima of x^(1/x)')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([1.35,1.5])
plt.grid(True)
plt.legend()

plt.show()
```
